# Test: POVM circuits

In this notebook we are testing the POVM circuits for two, three and four elements.

```python
import sys 
sys.path.append('../../')

import itertools
import numpy as np
from qnn.quantum_neural_networks import StateDiscriminativeQuantumNeuralNetworks as nnd
```

### 2-elements POVM circuit

```python
# Parameters
n = 2
th_u, fi_u, lam_u = [0], [0], [0]
th1, th2  = [0], [np.pi]
th_v1, th_v2 = [0], [0]
fi_v1, fi_v2 = [0], [0]
lam_v1, lam_v2 = [0],  [0]
```

```python
circuit_2_outcomes = nnd.get_n_element_povm(n, 
                                            th_u, fi_u, lam_u, 
                                            th1, th2, 
                                            th_v1, th_v2, 
                                            fi_v1, fi_v2, 
                                            lam_v1, lam_v2)
```

```python
circuit_2_outcomes.draw(output='mpl')
```

### 3-elements POVM circuit

```python
circuit_4_outcomes = nnd.get_n_element_povm(n, 
                                            th_u, fi_u, lam_u, 
                                            th1, th2,
                                            th_v1, th_v2, 
                                            fi_v1, fi_v2, 
                                            lam_v1, lam_v2)
```

```python
circuit_4_outcomes.draw(output='mpl')
```
