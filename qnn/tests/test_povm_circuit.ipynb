<span style="font-family:Huawen Caiyun; font-size:1.2em;">In this notebook we test the 2, 3 and 4 elements POVM circuits</span>

```python
import sys 
sys.path.append('../')

import itertools
import numpy as np
from quantum_neural_networks import StateDiscriminativeQuantumNeuralNetworks as nnd
```

<span style="font-family:Huawen Caiyun; font-size:1.4em;">2-elements POVM circuit</span>

```python
# Parameters
n = 2
th_u, fi_u, lam_u = [0], [0], [0]
th1, th2  = [0], [np.pi]
th_v1, th_v2 = [0], [0]
fi_v1, fi_v2 = [0], [0]
lam_v1, lam_v2 = [0],  [0]
```

```python
circuit_2_outcomes = nnd.get_n_element_povm( n, 
                                            th_u, fi_u, lam_u, 
                                            th1, th2, 
                                            th_v1, th_v2, 
                                            fi_v1, fi_v2, 
                                            lam_v1, lam_v2 )
```

```python
circuit_2_outcomes.draw(output='mpl')
```
